# print_000
### optimiertes Signal der Pickup Probe

```python
data_0 = np.loadtxt('../data/print_000.csv', delimiter=',', skiprows=2)
t_0 = data_0[:,0]*10**6
V_0 = data_0[:,1]

figs, ax = plt.subplots(figsize=(8,6))
ax.errorbar(t_0, V_0, label='Envelope', lw=1.5)
plt.grid(which='minor')
plt.grid(which='major')
plt.xlabel(r'$t$ / $\mathrm{\mu s}$')
plt.ylabel(r'$U$ / V')
plt.xlim(-2.5,7.5)
plt.tight_layout
plt.legend()
plt.savefig('../LateX/plots/optimized_signal.pdf')
plt.show()
```

# print_001
### FID-Signal eines $\pi/2$-Pulses

```python
data_1 = np.loadtxt('../data/print_001.csv', delimiter=',', skiprows=2)
t_1 = data_1[:,0]*10**3
V_1 = data_1[:,1]
abfall = max(V_1)*np.exp(-1)
#abfall1 = [abfall]*len(t_1)
figs, ax = plt.subplots()
ax.errorbar(t_1, V_1, color='g', label='Envelope', lw=1)
ax.set_xticks(np.arange(-6, t_1.max() +1, 2.0))
plt.grid(which='minor')
plt.grid(which='major')
plt.xlabel(r'$t$ / ms')
plt.ylabel(r'$U$ / V')
#plt.xlim(-2.5,7.5)
ax.axhline(y=abfall, color='r', linestyle='dashed', label='1/e Abfall')
plt.tight_layout()
plt.legend()
plt.savefig('../LateX/plots/FID.pdf')
plt.show()
```
